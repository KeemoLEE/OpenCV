## 그림 불러오기
> cv2.imread("파일이름")<br>
> cv2.imread("파일경로/파일이름")<br>
> cv2.imread("파일경로/파일이름", 파일읽는 방식)<br>
* 불러온 결과물은 3차원 배열로 저장됨<br>
* 파일 읽는 방식 : flags 값<br>
cv2.IMREAD_COLOR(default), 1 : RGB 3 채널 color 이미지<br>
cv2.IMREAD_GRAYSCALE, 0 : gray scale 이미지<br>
cv2.IMREAD_UNCHANGED, -1 : 원본이미지 그대로(투명한 png파일, alpha channel 등 가능)<br>
이외 : ANYDEPTH, ANYCOLOR.....

```python
img = cv2.imread("dog2.jpg")
img.shape
```

## 그림창 생성(생략 가능)
> cv2.namedWindow('창이름')<br>
> cv2.namedWindow('창이름', cv2.WINDOW_NORMAL)<br> 
> cv2.namedWindow('창이름', cv2.WINDOW_AUTOSIZE)<br> 
* flags 값<br>
default : AUTOSIZE 기본 적용됨<br>
cv2.WINDOW_NORMAL : 원본이미지 크기의 창 생성(창크기 조절 가능), 이미지 클 때 사용<br>
cv2.WINDOW_AUTOSIZE : 원본이미지 크기로 고정하여 창생성<br><br>
## 그림 출력하기
> cv2.imshow('창이름', 출력할이미지)<br>
> cv2.waitKey(시간)<br>
* 시간은 ms 단위이며 0은 키입력을 무한정 기다림<br>
* 키가 눌려지면 해당 키의 아스키코드 값 반환(시간동안 키 입력 없으면 -1 반환)
## 그림창 종료
> cv2.destroyWindow('창이름') : 해당 창 종료<br>
> cv2.destroyAllWindows() : 모든 창 종료

# **2. 도형, 글자 삽입**
## 선 그리기
> cv2.line(img, start, end, color, thickness)
- img : 그림 그릴 이미지 파일
- start : 선이 시작되는 좌표 ( x, y )
- end : 선이 끝나는 좌표 ( x, y )
- color : BGR 순서 색상값 (0,0,0) ~ (255,255,255)
- thickness : 선 두께, pixel 값(정수)

```python
import cv2
img=cv2.imread('duck.jpg')
print(img.shape)
img = cv2.line(img,(50,260),(220,260),(0,0,255),3)
cv2.imshow('line',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 원 그리기
> cv2.circle(img, center, radian, color, thickness)
- center : 원의 중심좌표, ( x, y )
- radian : 원의 반지름, pixel 값(정수)
- thickness : 선 두께, -1이면 원의 속을 채움

```python
import cv2
img=cv2.imread('duck.jpg')
print(img.shape)
img = cv2.circle(img, (410,250), 65, (0,0,255))
cv2.imshow('line',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 사각형 그리기
> cv2.rectangle(img, start, end, color, thickness)
- start : 사각형 시작좌표
- end : 사각형 종료좌표
- thickness : 선 두께, -1이면 원의 속을 채움

```python
import cv2
img=cv2.imread('duck.jpg')
print(img.shape)
img = cv2.rectangle(img, (360,220), (470,280), (0,0,255), 1)
cv2.imshow('line',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 글자
> cv2.putText(img, text, org, font, fontScale, color, thickness)
- text : 문자열
- org : 문자열의 위치(문자열의 왼쪽 아래 끝, ( x, y ))
- font : 폰트 타입
FONT_HERSHEY_(SIMPLEX, PLAIN, DUPLEX, COMPLEX, TRIPLEX, COMPLEX_SMALL, SCRIPT_SIMPLEX, SCRIPT_COMPLEX)
FONT_ITALIC
- fontScale : 폰트 크기(정수)

```python
import cv2
img=cv2.imread('duck.jpg')
print(img.shape)
img = cv2.putText(img,'OpenCV!',(50,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(255,255,255),3)
cv2.imshow('line',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **3. 영상 출력**
## 컴퓨터에서 영상이란?
사진(frame)들의 모임<br>
* 애니메이션 효과 : 사람의 눈은 1초에 24장 이상의 사진을 보여주면 움직이는 것처럼 보인다.
## 비디오 객체 생성
> cv2.VideoCapture('경로/영상파일')<br>
> cv2.VideoCapture(숫자)<br>
* 숫자는 컴퓨터에 있는 웹캠 순서를 의미, 0,1,2....
## 프레임 읽어오기
> ret, frame = 비디오객체.read()
* 영상 프레임을 읽었다면 ret= True, 실패했다면 ret= False
* 프레임은 frame 변수에 저장
## 프레임 출력
> imshow('창이름', frame)
## 비디오 객체 닫기
> 객체명.release()
## 창 닫기
> cv2.destroyAllWindows()

```python
import cv2
cam=cv2.VideoCapture(0)
while cv2.waitKey(33)<0 :
    ret, frame = cam.read()
    cv2.imshow('video',frame)
cam.release()
cv2.destroyAllWindows()
```
